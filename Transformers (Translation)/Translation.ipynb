```python
from transformers import pipeline
```

```python
translator = pipeline("translation", model="t5-large")
```

```python
prefixes = {'en-fr':'translate English to French:',
            'en-gr':'translate English to German:'}
```

```python
def translate_multilanguage_pipeline(text, lang):
    prefix = prefixes[lang]
    print(translator(prefix + text))
```

```python
translate_multilanguage_pipeline('To be, or not to be: that is the question.', 'en-fr')
```

```python
translate_multilanguage_pipeline('To be, or not to be: that is the question.', 'en-gr')
```
